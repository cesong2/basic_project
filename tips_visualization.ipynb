```python
sns.distplot(df['total_bill'])
```

```python
a = df['total_bill']
mean = a.mean()
median = np.median(a)
mode=a.mode()
```

```python
sns.distplot(a, hist=False)
plt.axvline(mean, color='r', label='mean')
plt.axvline(median, color='b', label='median')
plt.axvline(mode[0], color='g', label='mode')
plt.legend()
plt.show()
```

```python
sns.boxplot(a, color='Purple')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, hue='day')
# sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex',
           fit_reg=False, markers=['^','s'],
           palette='ocean', row = 'sex', col='smoker')
plt.show()
```

```python
sns.stripplot(x ='day', y='total_bill', data=df,
              jitter=False)
plt.grid()
plt.axhline(20, color='black')
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data=df, hue='sex')
plt.grid()
plt.axhline(20, color='black')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.figure(figsize=(5,5))
plt.show()
```

```python
sns.countplot(x='day', data=df, hue='sex')
plt.show()
```
